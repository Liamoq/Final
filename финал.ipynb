```python
import time

class Flight:
      def __init__(self,direction,date,date_1,date_2,price):
        self.direction = direction
        self.date = date
        self.date_1 = date_1
        self.date_2 = date_2 
        self.price = price

      def __str__(self):
        return (f"направление {self.direction}, дата {self.date}, дата {self.date_1}, дата {self.date_2}, цена {self.price}")

class Travel_Agencies:
        def __init__(self,name):
            self.name = name
            
        
        def __str__(self):
            return (f"Тур агенство: {self.name}")

        def print_flight(self,flight):
            print(f"доступные рейсы:\n")
            for i in flight:
                print(i)       


class Sum_Direction:

        def __init__(self,name,date):
            self.name = name
            self.date = date

        def __str__(self):
            return (f"{self.name}, {self.date}") 
    
        def p_user(self, list_user, flight):
            b = 0
            for i in flight:
                if (list_user.name==i.direction) and ((list_user.date == i.date) or (list_user.date == i.date_1) or (list_user.date == i.date_2)):
                    b += i.price
                    print(f"Итоговая сумма:{b}")
                    break
            else:
                print(f"Такого рейса нет\n")
        
        def ticket(self,user,list_user,travel):
            print(f"Выбронный рейс компании {travel}, {list_user} пользователя {user}")
            time.sleep(2)
            print(f"статус бронирование билета {user}:\n")
            time.sleep(2) 
            print(f"пользователь {user}, бронь оформлена\n")
            time.sleep(2)
            print(f"пользователь {user}, бронь прошла успешно")


     
user = "Человек-паук"

flight = [Flight("Африка","1.01.2025","9.05.2025","20.05.25",500000),
          Flight("Мальдивы","12.01.2025","4.09.2025","12.03.2025",150000000),
          Flight("Казахстан","2.01.2025","18.10.1025","8.02.2025",15000)]

travel = Travel_Agencies("Улет") 
travel.print_flight(flight)


list_user = Sum_Direction("Мальдивы","12.01.2025")
list_user.p_user(list_user,flight)
list_user.ticket(user,list_user,travel)
```
